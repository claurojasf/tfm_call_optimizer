```python
import ffmpeg
import os  # Para manejar archivos y directorios
import librosa  # Para cargar y analizar archivos de audio
import shutil
from pydub.utils import mediainfo
from pydub import AudioSegment
import shutil
from pydub.silence import split_on_silence
from pydub.effects import normalize
```

```python

# Ruta de la carpeta de audios
audio_folder = "C:/Users/carol/tfm_call_optimizer/audios/Audios_DEMO"

# Verifica si la carpeta existe
if os.path.exists(audio_folder):
    print("La carpeta existe.")
    print("Archivos en la carpeta:")

    # Inicializa el contador para generar nombres de archivos únicos
    id_archivo = 1

    # Obtiene la lista de archivos en la carpeta
    archivos = os.listdir(audio_folder)

    # Verifica si hay archivos en la carpeta
    if archivos:
        # Recorre todos los archivos en la carpeta
        for filename in archivos:
            # Muestra el nombre de los archivos para asegurarse de que los estamos detectando
            print(f"Archivo encontrado: {filename}")

            # Verifica si el archivo es un archivo de audio (MP3 o WAV)
            if filename.lower().endswith((".mp3", ".wav")):
                # Generar el nuevo nombre con un ID único (por ejemplo: call0001.mp3 o call0001.wav)
                extension = filename.split('.')[-1]
                nuevo_nombre = f"call_{id_archivo:04d}.{extension}"
                
                # Ruta completa del archivo original y la ruta con el nuevo nombre
                archivo_original = os.path.join(audio_folder, filename)
                archivo_renombrado = os.path.join(audio_folder, nuevo_nombre)

                # Renombrar el archivo
                try:
                    os.rename(archivo_original, archivo_renombrado)
                    print(f"Archivo renombrado: {filename} -> {nuevo_nombre}")
                    id_archivo += 1  # Incrementar el ID para el siguiente archivo
                except Exception as e:
                    print(f"Error al renombrar el archivo {filename}: {e}")
    else:
        print("No se encontraron archivos en la carpeta.")

else:
    print("La carpeta no existe.")
```

```python

# Ruta de la carpeta con los archivos .WAV
carpeta = 'C:/Users/carol/tfm_call_optimizer/audios/Audios_demo'

# Ruta de la carpeta donde se guardarán los archivos .MP3 (ya creada manualmente)
carpeta_mp3 = 'C:/Users/carol/tfm_call_optimizer/audios/Audios_mp3_DEMO' 

# Verificar si hay archivos .WAV en la carpeta, sin importar mayúsculas/minúsculas
archivos_wav = [archivo for archivo in os.listdir(carpeta) if archivo.lower().endswith('.wav')]

if not archivos_wav:
    print("No se encontraron archivos .WAV en la carpeta.")
else:
    print(f"Archivos .WAV encontrados: {archivos_wav}")

# Iterar sobre los archivos en la carpeta
for archivo in archivos_wav:
    # Ruta completa del archivo WAV
    archivo_wav = os.path.join(carpeta, archivo)

    # Crear el nombre para el archivo MP3 en la carpeta Audios_mp3
    archivo_mp3 = os.path.join(carpeta_mp3, archivo.replace('.WAV', '.mp3').replace('.wav', '.mp3'))

    # Asegurarse de que la ruta se maneje correctamente (evitar problemas con espacios)
    archivo_wav = os.path.abspath(archivo_wav)  # Convertir la ruta a su forma absoluta
    archivo_mp3 = os.path.abspath(archivo_mp3)

    # Verificar si el archivo WAV se está convirtiendo
    print(f"Convirtiendo '{archivo_wav}' a '{archivo_mp3}'...")

    # Convertir de WAV a MP3 utilizando ffmpeg
    try:
        ffmpeg.input(archivo_wav).output(archivo_mp3).run()
        print(f"Convertido: {archivo} a MP3")
    except ffmpeg.Error as e:
        print(f"Error al convertir {archivo}: {e}")

# Mostrar los archivos MP3 convertidos
print("\nArchivos MP3 convertidos:")
mp3_files = [archivo for archivo in os.listdir(carpeta_mp3) if archivo.lower().endswith('.mp3')]

if mp3_files:
    for archivo in mp3_files:
        print(archivo)
else:
    print("No se encontraron archivos MP3 en la carpeta.")
```

```python

# Definir rutas y umbrales
carpeta_mp3 = 'C:/Users/carol/tfm_call_optimizer/audios/Audios_mp3_DEMO'
carpeta_respaldo = 'C:/Users/carol/tfm_call_optimizer/audios/Audios_respaldo'  # Nueva carpeta de respaldo
umbral_minimo = 50  # 50 segundos
umbral_maximo = 10 * 60  # 5 minutos = 300 segundos

# Crear la carpeta de respaldo si no existe
if not os.path.exists(carpeta_respaldo):
    os.makedirs(carpeta_respaldo)

# Función para obtener la duración del archivo de audio usando pydub
def obtener_duracion(archivo):
    audio_path = os.path.join(carpeta_mp3, archivo)
    try:
        # Obtener información sobre el archivo de audio
        info = mediainfo(audio_path)
        duracion = float(info['duration'])  # Duración en segundos
        return duracion
    except Exception as e:
        print(f"Error al cargar el archivo {archivo}: {e}")
        return None

# Función para filtrar archivos por duración
def filtrar_por_duracion(archivos_mp3, umbral_minimo, umbral_maximo):
    archivos_a_respaldo = []  # Lista para almacenar archivos a mover
    archivos_validos = []  # Lista para almacenar archivos que cumplen el rango de duración
    for archivo in archivos_mp3:
        duracion = obtener_duracion(archivo)
        if duracion is not None:
            print(f"Duración de {archivo}: {duracion:.2f} segundos")
            # Verificar si el archivo está fuera del rango
            if duracion < umbral_minimo or duracion > umbral_maximo:
                archivos_a_respaldo.append(archivo)  # Agregar archivo a mover
            else:
                archivos_validos.append(archivo)  # Agregar archivo válido
        else:
            print(f"No se pudo obtener la duración de {archivo}.")
    
    # Mover los archivos fuera del rango a la carpeta de respaldo
    for archivo in archivos_a_respaldo:
        archivo_a_mover = os.path.join(carpeta_mp3, archivo)
        try:
            destino = os.path.join(carpeta_respaldo, archivo)
            shutil.move(archivo_a_mover, destino)  # Mover archivo
            print(f"Movido: {archivo} a la carpeta de respaldo")
        except Exception as e:
            print(f"Error al mover {archivo}: {e}")

    return archivos_validos  # Retornar los archivos que cumplen el rango de duración

# Función para eliminar silencios de un archivo de audio
def eliminar_silencio(audio_path, silencio_umbral=-40, min_silencio=1000):
    try:
        # Cargar el archivo de audio con pydub
        audio = AudioSegment.from_mp3(audio_path)

        # Usar split_on_silence para dividir el audio en segmentos de silencio
        segmentos_silencio = split_on_silence(audio, 
                                              min_silence_len=min_silencio,  # Mínima longitud de silencio en milisegundos
                                              silence_thresh=silencio_umbral)  # Umbral de silencio en dBFS (decibelios)

        # Unir los segmentos de audio sin silencio
        audio_sin_silencio = AudioSegment.empty()
        for segmento in segmentos_silencio:
            audio_sin_silencio += segmento

        # Guardar el archivo procesado, reemplazando el original
        audio_sin_silencio.export(audio_path, format="mp3")
        print(f"Silencio eliminado y archivo guardado: {audio_path}")
    except Exception as e:
        print(f"Error al procesar el archivo {audio_path}: {e}")

# Función principal para cargar los archivos, realizar el filtrado y eliminar silencios
def procesar_archivos():
    # Filtrar archivos MP3
    archivos_mp3 = [archivo for archivo in os.listdir(carpeta_mp3) if archivo.lower().endswith('.mp3')]
    
    if not archivos_mp3:
        print("No se encontraron archivos MP3 en la carpeta.")
        return
    
    # Filtrar y obtener los archivos dentro del rango de duración
    archivos_validos = filtrar_por_duracion(archivos_mp3, umbral_minimo, umbral_maximo)

    # Eliminar silencios solo en los archivos válidos
    for archivo in archivos_validos:
        audio_path = os.path.join(carpeta_mp3, archivo)
        eliminar_silencio(audio_path)  # Eliminar silencios y reemplazar el archivo original

if __name__ == "__main__":
    procesar_archivos()
```
